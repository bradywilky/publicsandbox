```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Define the URL of the League of Legends Fandom item page
url = "https://leagueoflegends.fandom.com/wiki/List_of_items"

# Send an HTTP GET request to the URL
response = requests.get(url)

if response.status_code == 200:
    # Parse the HTML content
    soup = BeautifulSoup(response.text, "html.parser")
    
    items = list()
    for i in soup.find_all('li'):
        for j in i.find_all('div', {'class': 'item-icon'}):
            if '5v5' in j.get('data-modes'):
                items.append(j.get('data-item'))
```

```python
item_urls = [i.replace(' ', '_').replace("'", "%27") for i in items]
item_urls
```

```python
t = list()
for i in item_urls:
    
    # Define the URL of the League of Legends Fandom item page
    url = f"https://leagueoflegends.fandom.com/wiki/{i}"

    # Send an HTTP GET request to the URL
    response = requests.get(url)

    if response.status_code == 200:
        # Parse the HTML content
        t.append(i)
```

```python
# Define the URL of the League of Legends Fandom item page
url = f"https://leagueoflegends.fandom.com/wiki/Doran%27s_Ring"

# Send an HTTP GET request to the URL
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

ds_list = ['ap', 'ad', 'hp', 'mp', 'ms', 'mp5', 'hp5']
for t in soup.findAll('div', {'class': 'pi-item pi-data pi-item-spacing pi-border-color'}):
    if t.get('data-source') in ds_list:
        info_tag = t.find('div')
        name = info_tag.contents[1].contents[0]
        stat = info_tag.contents[0]

        print(name, stat)
    else:
        print(t.get('data-source'))
```
